```python
# Stock Price Prediction Challenge Solution
# This notebook provides a complete workflow for predicting stock prices 5 days in the future

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split, TimeSeriesSplit, GridSearchCV
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.linear_model import LinearRegression
import xgboost as xgb
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.tsa.stattools import adfuller
import warnings

warnings.filterwarnings('ignore')

# 1. Data Loading and Initial Exploration
# Assuming data is in a CSV file named 'stock_data.csv'
def load_and_explore_data(file_path='/content/drive/MyDrive/Task4/question4-stock-data.csv'):
    """
    Load the stock data and perform initial exploration
    """
    print("Loading data...")
    # Load data
    df = pd.read_csv(file_path, index_col=0)

    # Convert date to datetime
    if 'Date' in df.columns:
        df['Date'] = pd.to_datetime(df['Date'])
        df.set_index('Date', inplace=True)

    # Display basic information
    print("\nDataset Overview:")
    print(f"Shape: {df.shape}")
    print("\nFirst few rows:")
    print(df.head())

    print("\nData types:")
    print(df.dtypes)

    print("\nSummary statistics:")
    print(df.describe())

    print("\nMissing values:")
    print(df.isnull().sum())

    return df

# 2. Comprehensive Exploratory Data Analysis
def perform_eda(df):
    """
    Perform comprehensive exploratory data analysis
    """
    print("\n--- Exploratory Data Analysis ---")

    # Plot the closing price over time
    plt.figure(figsize=(14, 7))
    plt.plot(df.index, df['Close'], label='Close Price')
    plt.title('Stock Close Price Over Time')
    plt.xlabel('Date')
    plt.ylabel('Price')
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.savefig('close_price_over_time.png')
    plt.show()

     # Distribution of closing prices
    plt.figure(figsize=(14, 7))
    plt.subplot(1, 2, 1)
    sns.histplot(df['Close'], kde=True)
    plt.title('Distribution of Close Prices')

    plt.subplot(1, 2, 2)
    # Reset the index before creating the boxplot to avoid duplicate labels:
    sns.boxplot(y=df.reset_index()['Close'])
    plt.title('Boxplot of Close Prices')
    plt.tight_layout()
    plt.savefig('close_price_distribution.png')
    plt.show()

    # Daily returns calculation and analysis
    df['Daily_Return'] = df['Close'].pct_change() * 100

    plt.figure(figsize=(14, 7))
    plt.subplot(1, 2, 1)
    plt.plot(df.index, df['Daily_Return'])
    plt.title('Daily Returns Over Time')
    plt.xlabel('Date')
    plt.ylabel('Return (%)')

    plt.subplot(1, 2, 2)
    sns.histplot(df['Daily_Return'].dropna(), kde=True)
    plt.title('Distribution of Daily Returns')
    plt.tight_layout()
    plt.savefig('daily_returns.png')
    plt.show()

    # Volatility (rolling standard deviation of returns)
    df['Volatility'] = df['Daily_Return'].rolling(window=21).std()  # 21 trading days = 1 month

    plt.figure(figsize=(14, 7))
    plt.plot(df.index, df['Volatility'])
    plt.title('21-Day Rolling Volatility')
    plt.xlabel('Date')
    plt.ylabel('Volatility')
    plt.grid(True)
    plt.tight_layout()
    plt.savefig('volatility.png')
    plt.show()

    # Volume analysis
    if 'Volume' in df.columns:
        plt.figure(figsize=(14, 7))
        plt.subplot(2, 1, 1)
        plt.plot(df.index, df['Volume'])
        plt.title('Trading Volume Over Time')
        plt.xlabel('Date')
        plt.ylabel('Volume')
        plt.grid(True)

        plt.subplot(2, 1, 2)
        plt.scatter(df['Daily_Return'], df['Volume'], alpha=0.5)
        plt.title('Volume vs Daily Returns')
        plt.xlabel('Daily Return (%)')
        plt.ylabel('Volume')
        plt.grid(True)
        plt.tight_layout()
        plt.savefig('volume_analysis.png')
        plt.show()

    # Correlation analysis
    plt.figure(figsize=(12, 10))
    corr_matrix = df.corr()
    sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f')
    plt.title('Correlation Matrix')
    plt.tight_layout()
    plt.savefig('correlation_matrix.png')
    plt.show()

    # Time series decomposition
    try:
        decomposition = seasonal_decompose(df['Close'], model='multiplicative', period=252)  # 252 trading days in a year

        plt.figure(figsize=(14, 12))
        plt.subplot(4, 1, 1)
        plt.plot(decomposition.observed)
        plt.title('Observed')
        plt.grid(True)

        plt.subplot(4, 1, 2)
        plt.plot(decomposition.trend)
        plt.title('Trend')
        plt.grid(True)

        plt.subplot(4, 1, 3)
        plt.plot(decomposition.seasonal)
        plt.title('Seasonality')
        plt.grid(True)

        plt.subplot(4, 1, 4)
        plt.plot(decomposition.resid)
        plt.title('Residuals')
        plt.grid(True)

        plt.tight_layout()
        plt.savefig('time_series_decomposition.png')
        plt.show()
    except:
        print("Could not perform time series decomposition. Ensure sufficient data points.")

    # Stationarity test
    print("\nPerforming Augmented Dickey-Fuller test for stationarity on Close price:")
    adf_result = adfuller(df['Close'].dropna())
    print(f'ADF Statistic: {adf_result[0]:.4f}')
    print(f'p-value: {adf_result[1]:.4f}')
    print('Critical Values:')
    for key, value in adf_result[4].items():
        print(f'\t{key}: {value:.4f}')

    if adf_result[1] > 0.05:
        print("The price series is non-stationary.")
    else:
        print("The price series is stationary.")

    # Test stationarity of returns
    print("\nPerforming Augmented Dickey-Fuller test for stationarity on Daily Returns:")
    adf_result = adfuller(df['Daily_Return'].dropna())
    print(f'ADF Statistic: {adf_result[0]:.4f}')
    print(f'p-value: {adf_result[1]:.4f}')

    if adf_result[1] > 0.05:
        print("The returns series is non-stationary.")
    else:
        print("The returns series is stationary.")

    return df

# 3. Feature Engineering
def engineer_features(df):
    """
    Create features from the time series data
    """
    print("\n--- Feature Engineering ---")

    # Make a copy to avoid modifying the original dataframe
    features_df = df.copy()

    # 1. Technical Indicators

    # Moving Averages
    for window in [5, 10, 20, 50, 100, 200]:
        features_df[f'MA_{window}'] = features_df['Close'].rolling(window=window).mean()  # Use 'Close' column

    # Exponential Moving Averages
    for window in [5, 10, 20, 50]:
        features_df[f'EMA_{window}'] = features_df['Close'].ewm(span=window, adjust=False).mean()

    # Moving Average Convergence Divergence (MACD)
    features_df['EMA_12'] = features_df['Close'].ewm(span=12, adjust=False).mean()
    features_df['EMA_26'] = features_df['Close'].ewm(span=26, adjust=False).mean()
    features_df['MACD'] = features_df['EMA_12'] - features_df['EMA_26']
    features_df['MACD_Signal'] = features_df['MACD'].ewm(span=9, adjust=False).mean()
    features_df['MACD_Hist'] = features_df['MACD'] - features_df['MACD_Signal']

    # Relative Strength Index (RSI)
    delta = features_df['Close'].diff()
    gain = delta.where(delta > 0, 0)
    loss = -delta.where(delta < 0, 0)

    avg_gain = gain.rolling(window=14).mean()
    avg_loss = loss.rolling(window=14).mean()

    rs = avg_gain / avg_loss
    features_df['RSI'] = 100 - (100 / (1 + rs))

    # Bollinger Bands
    features_df['20MA'] = features_df['Close'].rolling(window=20).mean()
    features_df['20SD'] = features_df['Close'].rolling(window=20).std()
    features_df['Upper_Band'] = features_df['20MA'] + (features_df['20SD'] * 2)
    features_df['Lower_Band'] = features_df['20MA'] - (features_df['20SD'] * 2)
    features_df['BB_Width'] = (features_df['Upper_Band'] - features_df['Lower_Band']) / features_df['20MA']

    # Average True Range (ATR)
    high_low = features_df['High'] - features_df['Low']
    high_close = np.abs(features_df['High'] - features_df['Close'].shift())
    low_close = np.abs(features_df['Low'] - features_df['Close'].shift())

    ranges = pd.concat([high_low, high_close, low_close], axis=1)
    true_range = np.max(ranges, axis=1)
    features_df['ATR'] = true_range.rolling(14).mean()

    # 2. Price momentum and rate of change
    for window in [1, 3, 5, 10, 20]:
        features_df[f'Price_Change_{window}'] = features_df['Close'].pct_change(periods=window) * 100  # Use 'Close' column

    # 3. Volatility features
    for window in [5, 10, 21, 63]:  # 1 week, 2 weeks, 1 month, 3 months
        features_df[f'Volatility_{window}'] = features_df['Daily_Return'].rolling(window=window).std()

    # 4. Volume indicators (if volume data is available)
    if 'Volume' in features_df.columns:
        # Volume moving averages
        for window in [5, 10, 20]:
            features_df[f'Volume_MA_{window}'] = features_df['Volume'].rolling(window=window).mean()

        # Volume rate of change
        features_df['Volume_Change_1'] = features_df['Volume'].pct_change() * 100

        # On-Balance Volume (OBV)
        features_df['OBV'] = np.where(features_df['Close'] > features_df['Close'].shift(1),
                                 features_df['Volume'],
                                 np.where(features_df['Close'] < features_df['Close'].shift(1),
                                         -features_df['Volume'], 0)).cumsum()

    # 5. Date-based features
    if isinstance(features_df.index, pd.DatetimeIndex):
        features_df['Day_of_Week'] = features_df.index.dayofweek
        features_df['Month'] = features_df.index.month
        features_df['Quarter'] = features_df.index.quarter
        features_df['Year'] = features_df.index.year
        features_df['Day_of_Year'] = features_df.index.dayofyear

    # 6. Target variable: Close price 5 days into the future
    features_df['Target'] = features_df['Close'].shift(-5)  # Use 'Close' column

    # Display newly created features
    print(f"Created {len(features_df.columns) - len(df.columns)} new features.")

    # Visualize some key features
    plt.figure(figsize=(14, 10))

    plt.subplot(2, 2, 1)
    plt.plot(features_df.index, features_df['Close'], label='Close Price')
    plt.plot(features_df.index, features_df['MA_50'], label='50-day MA')
    plt.plot(features_df.index, features_df['MA_200'], label='200-day MA')
    plt.title('Moving Averages')
    plt.legend()
    plt.grid(True)

    plt.subplot(2, 2, 2)
    plt.plot(features_df.index, features_df['RSI'], label='RSI')
    plt.axhline(y=70, color='r', linestyle='--')
    plt.axhline(y=30, color='g', linestyle='--')
    plt.title('Relative Strength Index (RSI)')
    plt.legend()
    plt.grid(True)

    plt.subplot(2, 2, 3)
    plt.plot(features_df.index, features_df['MACD'], label='MACD')
    plt.plot(features_df.index, features_df['MACD_Signal'], label='Signal Line')
    plt.bar(features_df.index, features_df['MACD_Hist'], alpha=0.3, label='MACD Histogram')
    plt.title('MACD')
    plt.legend()
    plt.grid(True)

    plt.subplot(2, 2, 4)
    plt.plot(features_df.index, features_df['Close'], label='Close Price')
    plt.plot(features_df.index, features_df['Upper_Band'], label='Upper Band')
    plt.plot(features_df.index, features_df['Lower_Band'], label='Lower Band')
    plt.title('Bollinger Bands')
    plt.legend()
    plt.grid(True)

    plt.tight_layout()
    plt.savefig('technical_indicators.png')
    plt.show()

    # Display correlation of features with target
    correlation_with_target = features_df.corr()['Target'].sort_values(ascending=False)
    print("\nTop 10 features correlated with target:")
    print(correlation_with_target.head(11))  # 11 because target will be included

    print("\nBottom 10 features correlated with target:")
    print(correlation_with_target.tail(10))

    return features_df

# 4. Prepare Data for Modeling
def prepare_for_modeling(features_df):
    """
    Prepare the data for modeling
    """
    print("\n--- Data Preparation for Modeling ---")

    # Drop rows with NaN values
    df_clean = features_df.dropna()
    print(f"Dataset shape after dropping NaN values: {df_clean.shape}")

    # Split features and target
    X = df_clean.drop('Target', axis=1)
    y = df_clean['Target']

    # Save the date index for later
    X_dates = X.index

    columns_to_drop = ['Open', 'High', 'Low', 'Close', 'Adj Close', 'Daily_Return', 'Volatility']  # Include all unnecessary columns

    # Feature importance based on correlation (for initial feature selection)
    correlation_with_target = features_df.corr()['Target'].abs().sort_values(ascending=False)

    # Select top features based on correlation, EXCLUDING columns to drop
    top_features = [feature for feature in correlation_with_target[1:26].index.tolist() if feature not in columns_to_drop]  # Exclude Target itself and take top 25
    print(f"\nTop features selected based on correlation (excluding columns to drop):")
    for i, feature in enumerate(top_features, 1):
        print(f"{i}. {feature}: {correlation_with_target[feature]:.4f}")

    X = X.drop(columns_to_drop, axis=1, errors='ignore')
    X_selected = X[top_features]

    # Split the data into training and validation sets using time series split
    # For time series data, we need to maintain the chronological order
    train_size = int(0.8 * len(X_selected))
    X_train, X_val = X_selected.iloc[:train_size], X_selected.iloc[train_size:]
    y_train, y_val = y.iloc[:train_size], y.iloc[train_size:]

    print(f"\nTraining set shape: {X_train.shape}")
    print(f"Validation set shape: {X_val.shape}")

    # --- Handling NaN and inf ---
    # Replace infinite values in X_train with NaN:
    X_train = X_train.replace([np.inf, -np.inf], np.nan)
    # Fill NaN values with the column mean in X_train:
    for column in X_train.columns[X_train.isnull().any()]:
        X_train[column] = X_train[column].fillna(X_train[column].mean())
    # Replace any remaining infinite values in X_train with 0 (or another suitable value):
    X_train = X_train.replace([np.inf, -np.inf], 0)

    # --- Scaling X_train ---
    scaler = StandardScaler()
    X_train_scaled = scaler.fit_transform(X_train)

    # --- Handling NaN and inf in X_val ---
    # Replace infinite values in X_val with NaN:
    X_val = X_val.replace([np.inf, -np.inf], np.nan)
    # Fill NaN values with the column mean in X_val:
    for column in X_val.columns[X_val.isnull().any()]:
        X_val[column] = X_val[column].fillna(X_val[column].mean())
    # Replace any remaining infinite values in X_val with 0 (or another suitable value):
    X_val = X_val.replace([np.inf, -np.inf], 0)

    # --- Scaling X_val ---
    X_val_scaled = scaler.transform(X_val)

    # Convert back to DataFrame for better interpretability
    X_train_scaled = pd.DataFrame(X_train_scaled, columns=X_train.columns, index=X_train.index)
    X_val_scaled = pd.DataFrame(X_val_scaled, columns=X_val.columns, index=X_val.index)

    return X_train_scaled, X_val_scaled, y_train, y_val, scaler, top_features

# 5. Model Development and Evaluation
def develop_and_evaluate_models(X_train, X_val, y_train, y_val, top_features):
    """
    Develop and evaluate multiple models
    """
    print("\n--- Model Development and Evaluation ---")

    # Initialize models to test
    models = {
        'Linear Regression': LinearRegression(),
        'Random Forest': RandomForestRegressor(n_estimators=100, random_state=42),
        'Gradient Boosting': GradientBoostingRegressor(n_estimators=100, random_state=42),
        'XGBoost': xgb.XGBRegressor(n_estimators=100, random_state=42)  # Use xgb.XGBRegressor
    }

    # Dictionary to store model performance
    model_performances = {}

    # Function to calculate directional accuracy
    def directional_accuracy(y_true, y_pred):
        y_true_direction = np.sign(y_true.pct_change())
        y_pred_direction = np.sign(y_pred.pct_change())
        return np.mean(y_true_direction == y_pred_direction)

    # Train and evaluate each model
    for name, model in models.items():
        print(f"\nTraining {name}...")
        model.fit(X_train, y_train)

        # Make predictions on validation set
        val_predictions = model.predict(X_val)

        # Calculate performance metrics
        rmse = np.sqrt(mean_squared_error(y_val, val_predictions))
        mae = mean_absolute_error(y_val, val_predictions)
        r2 = r2_score(y_val, val_predictions)

        # Convert predictions to pandas Series for directional accuracy
        val_pred_series = pd.Series(val_predictions, index=y_val.index)
        dir_accuracy = directional_accuracy(y_val, val_pred_series)

        # Store performance
        model_performances[name] = {
            'RMSE': rmse,
            'MAE': mae,
            'R-squared': r2,
            'Directional Accuracy': dir_accuracy,
            'model': model
        }

        # Print results
        print(f"{name} Performance:")
        print(f"RMSE: {rmse:.4f}")
        print(f"MAE: {mae:.4f}")
        print(f"R-squared: {r2:.4f}")
        print(f"Directional Accuracy: {dir_accuracy:.4f}")

        # Plot actual vs predicted
        plt.figure(figsize=(14, 7))
        plt.plot(y_val.index, y_val, label='Actual')
        plt.plot(y_val.index, val_predictions, label='Predicted')
        plt.title(f'{name} - Actual vs Predicted Stock Prices')
        plt.xlabel('Date')
        plt.ylabel('Price')
        plt.legend()
        plt.grid(True)
        plt.savefig(f'{name}_predictions.png')
        plt.show()

        # Feature importance (if applicable)
        if hasattr(model, 'feature_importances_'):
            feature_importance = pd.DataFrame({
                'Feature': top_features,
                'Importance': model.feature_importances_
            }).sort_values('Importance', ascending=False)

            plt.figure(figsize=(10, 8))
            sns.barplot(x='Importance', y='Feature', data=feature_importance.head(15))
            plt.title(f'Top 15 Feature Importances - {name}')
            plt.tight_layout()
            plt.savefig(f'{name}_feature_importance.png')
            plt.show()

            print("\nTop 10 most important features:")
            print(feature_importance.head(10))

    # Determine the best model based on RMSE
    best_model_name = min(model_performances, key=lambda x: model_performances[x]['RMSE'])
    print(f"\nBest performing model: {best_model_name}")
    print(f"RMSE: {model_performances[best_model_name]['RMSE']:.4f}")
    print(f"Directional Accuracy: {model_performances[best_model_name]['Directional Accuracy']:.4f}")

    # Compare all models
    plt.figure(figsize=(12, 6))
    model_names = list(model_performances.keys())
    rmse_values = [model_performances[model]['RMSE'] for model in model_names]
    dir_acc_values = [model_performances[model]['Directional Accuracy'] for model in model_names]

    plt.subplot(1, 2, 1)
    plt.bar(model_names, rmse_values)
    plt.title('RMSE Comparison')
    plt.ylabel('RMSE (lower is better)')
    plt.xticks(rotation=45)

    plt.subplot(1, 2, 2)
    plt.bar(model_names, dir_acc_values)
    plt.title('Directional Accuracy Comparison')
    plt.ylabel('Directional Accuracy (higher is better)')
    plt.xticks(rotation=45)

    plt.tight_layout()
    plt.savefig('model_comparison.png')
    plt.show()

    return model_performances, best_model_name

# 6. Simulated Trading Evaluation
def evaluate_trading_strategy(y_val, model_predictions, initial_capital=10000):
    """
    Evaluate model performance using a simple trading strategy
    """
    print("\n--- Simulated Trading Evaluation ---")

    # Create a dataframe for the trading simulation
    trades_df = pd.DataFrame({
        'Actual_Price': y_val,
        'Predicted_Price': model_predictions
    })

    # Calculate predicted price changes and actual price changes
    trades_df['Predicted_Change'] = trades_df['Predicted_Price'].pct_change()
    trades_df['Actual_Change'] = trades_df['Actual_Price'].pct_change()

    # Trading strategy: Buy when predicted price is higher than current, sell when lower
    trades_df['Position'] = np.sign(trades_df['Predicted_Change'])
    trades_df['Position'] = trades_df['Position'].shift(1)  # Shift by 1 because we make decision based on previous prediction

    # Calculate returns: Position * Actual_Change (returns are realized next day)
    trades_df['Strategy_Return'] = trades_df['Position'] * trades_df['Actual_Change']

    # Buy and hold returns
    trades_df['Buy_Hold_Return'] = trades_df['Actual_Change']

    # Drop NaN values
    trades_df = trades_df.dropna()

    # Calculate cumulative returns
    trades_df['Cumulative_Strategy_Return'] = (1 + trades_df['Strategy_Return']).cumprod()
    trades_df['Cumulative_Buy_Hold_Return'] = (1 + trades_df['Buy_Hold_Return']).cumprod()

    # Calculate strategy performance metrics
    total_strategy_return = trades_df['Cumulative_Strategy_Return'].iloc[-1] - 1
    total_buy_hold_return = trades_df['Cumulative_Buy_Hold_Return'].iloc[-1] - 1

    annualized_factor = 252 / len(trades_df)  # 252 trading days in a year
    annualized_strategy_return = (1 + total_strategy_return) ** annualized_factor - 1
    annualized_buy_hold_return = (1 + total_buy_hold_return) ** annualized_factor - 1

    sharpe_ratio = (trades_df['Strategy_Return'].mean() / trades_df['Strategy_Return'].std()) * (252 ** 0.5)

    # Calculate drawdowns
    strategy_peak = trades_df['Cumulative_Strategy_Return'].cummax()
    strategy_drawdown = (trades_df['Cumulative_Strategy_Return'] / strategy_peak) - 1
    max_drawdown = strategy_drawdown.min()

    # Print performance metrics
    print(f"Total Strategy Return: {total_strategy_return:.4f} ({total_strategy_return*100:.2f}%)")
    print(f"Total Buy & Hold Return: {total_buy_hold_return:.4f} ({total_buy_hold_return*100:.2f}%)")
    print(f"Annualized Strategy Return: {annualized_strategy_return:.4f} ({annualized_strategy_return*100:.2f}%)")
    print(f"Annualized Buy & Hold Return: {annualized_buy_hold_return:.4f} ({annualized_buy_hold_return*100:.2f}%)")
    print(f"Sharpe Ratio: {sharpe_ratio:.4f}")
    print(f"Maximum Drawdown: {max_drawdown:.4f} ({max_drawdown*100:.2f}%)")

    # Calculate portfolio values
    strategy_portfolio = initial_capital * trades_df['Cumulative_Strategy_Return']
    buyhold_portfolio = initial_capital * trades_df['Cumulative_Buy_Hold_Return']

    # Plot performance comparison
    plt.figure(figsize=(14, 7))
    plt.plot(trades_df.index, strategy_portfolio, label=f'Trading Strategy (${strategy_portfolio.iloc[-1]:.2f})')
    plt.plot(trades_df.index, buyhold_portfolio, label=f'Buy & Hold (${buyhold_portfolio.iloc[-1]:.2f})')
    plt.title(f'Performance Comparison - Initial Capital: ${initial_capital}')
    plt.xlabel('Date')
    plt.ylabel('Portfolio Value ($)')
    plt.legend()
    plt.grid(True)
    plt.savefig('trading_strategy_performance.png')
    plt.show()

    # Plot drawdowns
    plt.figure(figsize=(14, 7))
    plt.plot(trades_df.index, strategy_drawdown * 100)
    plt.title('Strategy Drawdowns')
    plt.xlabel('Date')
    plt.ylabel('Drawdown (%)')
    plt.fill_between(trades_df.index, strategy_drawdown * 100, 0, alpha=0.3, color='red')
    plt.grid(True)
    plt.savefig('strategy_drawdowns.png')
    plt.show()

    return trades_df, {
        'Total Return': total_strategy_return,
        'Annualized Return': annualized_strategy_return,
        'Sharpe Ratio': sharpe_ratio,
        'Max Drawdown': max_drawdown,
        'Buy & Hold Return': total_buy_hold_return
    }

# 7. Model Hyperparameter Tuning
def tune_best_model(X_train, y_train, X_val, y_val, best_model_name):
    """
    Perform hyperparameter tuning for the best model
    """
    print(f"\n--- Hyperparameter Tuning for {best_model_name} ---")

    # Define parameter grids for each model type
    param_grids = {
        'Linear Regression': {},  # Linear regression doesn't have hyperparameters to tune

        'Random Forest': {
            'n_estimators': [50, 100, 200],
            'max_depth': [None, 10, 20, 30],
            'min_samples_split': [2, 5, 10],
            'min_samples_leaf': [1, 2, 4]
        },

        'Gradient Boosting': {
            'n_estimators': [50, 100, 200],
            'learning_rate': [0.01, 0.05, 0.1],
            'max_depth': [3, 5, 7],
            'min_samples_split': [2, 5],
            'min_samples_leaf': [1, 2]
        },

        'XGBoost': {
            'n_estimators': [50, 100, 200],
            'learning_rate': [0.01, 0.05, 0.1],
            'max_depth': [3, 5, 7],
            'min_child_weight': [1, 3, 5],
            'subsample': [0.8, 0.9, 1.0]
        }
    }

    # If the best model is Linear Regression, we don't need to tune hyperparameters
    if best_model_name == 'Linear Regression':
        print("Linear Regression doesn't have hyperparameters to tune. Skipping this step.")
        return None

     # Initialize model based on best_model_name
    if best_model_name == 'Random Forest':
        model = RandomForestRegressor(random_state=42)
    elif best_model_name == 'Gradient Boosting':
        model = GradientBoostingRegressor(random_state=42)
    elif best_model_name == 'XGBoost':
        model = xgb.XGBRegressor(random_state=42)

    # Get parameter grid for the best model
    param_grid = param_grids[best_model_name]

    # Time series cross-validation
    tscv = TimeSeriesSplit(n_splits=5)

    # Initialize grid search
    grid_search = GridSearchCV(
        estimator=model,
        param_grid=param_grid,
        cv=tscv,
        scoring='neg_mean_squared_error',
        n_jobs=-1,
        verbose=1
    )

    # Fit grid search
    grid_search.fit(X_train, y_train)

    # Print best parameters and score
    print(f"Best parameters: {grid_search.best_params_}")
    print(f"Best RMSE: {np.sqrt(-grid_search.best_score_):.4f}")

    # Evaluate tuned model on validation set
    tuned_model = grid_search.best_estimator_
    val_predictions = tuned_model.predict(X_val)
    tuned_rmse = np.sqrt(mean_squared_error(y_val, val_predictions))
    tuned_r2 = r2_score(y_val, val_predictions)

    print(f"Validation RMSE with tuned model: {tuned_rmse:.4f}")
    print(f"Validation R-squared with tuned model: {tuned_r2:.4f}")

    # Compare with untuned model
    plt.figure(figsize=(14, 7))
    plt.plot(y_val.index, y_val, label='Actual')
    plt.plot(y_val.index, val_predictions, label='Tuned Model Predictions')
    plt.title(f'Tuned {best_model_name} - Actual vs Predicted Stock Prices')
    plt.xlabel('Date')
    plt.ylabel('Price')
    plt.legend()
    plt.grid(True)
    plt.savefig(f'tuned_{best_model_name}_predictions.png')
    plt.show()

    return tuned_model

# 8. Generate Predictions for Test Period
def generate_test_predictions(model, features_df, scaler, top_features):
    """
    Generate predictions for the test period
    """
    print("\n--- Generating Test Predictions ---")

    # Create a copy of the features dataframe
    test_df = features_df.copy()

    # For demonstration, we'll use the last 30 days as our "test" period
    test_period = test_df.tail(30)
    print(f"Test period: {test_period.index[0]} to {test_period.index[-1]}")

    # Prepare test features
    # Ensure that top_features are actually present in the test_period DataFrame
    available_features = [f for f in top_features if f in test_period.columns]
    X_test = test_period[available_features]

    # Scale the features
    X_test_scaled = scaler.transform(X_test)
    X_test_scaled = pd.DataFrame(X_test_scaled, columns=X_test.columns, index=X_test.index)

    # Generate predictions
    test_predictions = model.predict(X_test_scaled)

    # Create DataFrame with predictions
    predictions_df = pd.DataFrame({
        'Date': X_test.index,
        'Predicted_Close': test_predictions
    })

    # For comparison, include actual prices if available
    # Instead of 'Close', use 'Target' which is shifted 'Close' values
    # Check if 'Target' column exists before accessing it
    if 'Target' in test_period.columns:
        actual_prices = test_period['Target']
        predictions_df['Actual_Close'] = actual_prices

        # Calculate prediction error
        predictions_df['Error'] = predictions_df['Actual_Close'] - predictions_df['Predicted_Close']
        predictions_df['Percent_Error'] = (predictions_df['Error'] / predictions_df['Actual_Close']) * 100

    # Display predictions
    print("\nTest Period Predictions:")
    print(predictions_df)

    # Save predictions to CSV
    predictions_df.to_csv('test_predictions.csv')
    print("Predictions saved to 'test_predictions.csv'")

    # Visualize predictions
    plt.figure(figsize=(14, 7))

    if 'Actual_Close' in predictions_df.columns:
        plt.plot(predictions_df['Date'], predictions_df['Actual_Close'], label='Actual')

    plt.plot(predictions_df['Date'], predictions_df['Predicted_Close'], label='Predicted')
    plt.title('Stock Price Predictions for Test Period')
    plt.xlabel('Date')
    plt.ylabel('Price')
    plt.legend()
    plt.grid(True)
    plt.savefig('test_predictions.png')
    plt.show()

    # Plot percent error
    if 'Percent_Error' in predictions_df.columns:
        plt.figure(figsize=(14, 7))
        plt.bar(predictions_df['Date'], predictions_df['Percent_Error'])
        plt.axhline(y=0, color='r', linestyle='-')
        plt.title('Prediction Error (%)')
        plt.xlabel('Date')
        plt.ylabel('Error (%)')
        plt.grid(True)
        plt.savefig('prediction_errors.png')
        plt.show()

    return predictions_df

# 9. Model Limitations and Improvement Strategies
def analyze_limitations():
    """
    Analyze model limitations and suggest improvements
    """
    limitations = {
        "Data Limitations": [
            "Limited historical data may not capture all market cycles",
            "Missing external factors like economic indicators, news sentiment, etc.",
            "No consideration of market regime changes (bull vs bear markets)",
            "Potential survivorship bias in the stock data",
            "No incorporation of sector-specific or market-wide factors"
        ],

        "Model Limitations": [
            "Linear models may not capture complex non-linear relationships",
            "Tree-based models might overfit to historical patterns",
            "No consideration of time-varying relationships between features",
            "Models don't account for unexpected events (e.g., earnings surprises, geopolitical events)",
            "Simple trading strategy doesn't account for transaction costs, slippage, or liquidity"
        ],

        "Improvement Strategies": [
            "Incorporate alternative data sources (news sentiment, social media, economic indicators)",
            "Implement more sophisticated models like LSTM or transformer networks for time series",
            "Use ensemble methods combining different model types",
            "Implement regime detection to adapt models to different market conditions",
            "Consider market microstructure for more realistic trading simulation",
            "Implement cross-sectional analysis across multiple stocks for relative value signals",
            "Develop more sophisticated trading strategies with risk management"
        ]
    }

    print("\n--- Model Limitations and Improvement Strategies ---")

    for category, items in limitations.items():
        print(f"\n{category}:")
        for i, item in enumerate(items, 1):
            print(f"{i}. {item}")

    return limitations

# 10. Main Function
def main():
    """
    Main function to run the entire workflow
    """
    print("======= STOCK PRICE PREDICTION CHALLENGE =======")

    # Step 1: Load and explore data
    df = load_and_explore_data()

    # Step 2: Perform EDA
    df = perform_eda(df)

    # Step 3: Engineer features
    features_df = engineer_features(df)

    # Step 4: Prepare data for modeling
    X_train, X_val, y_train, y_val, scaler, top_features = prepare_for_modeling(features_df)

    # Step 5: Develop and evaluate models
    model_performances, best_model_name = develop_and_evaluate_models(X_train, X_val, y_train, y_val, top_features)

    # Step 6: Evaluate trading strategy
    best_model = model_performances[best_model_name]['model']
    val_predictions = best_model.predict(X_val)
    trades_df, strategy_metrics = evaluate_trading_strategy(y_val, val_predictions)

    # Step 7: Tune the best model
    tuned_model = tune_best_model(X_train, y_train, X_val, y_val, best_model_name)

    # Use tuned model if available, otherwise use the original best model
    final_model = tuned_model if tuned_model is not None else best_model

    # Step 8: Generate predictions for test period
    predictions_df = generate_test_predictions(final_model, features_df, scaler, top_features)

    # Step 9: Analyze limitations
    limitations = analyze_limitations()

    print("\n======= STOCK PRICE PREDICTION CHALLENGE COMPLETED =======")

if __name__ == "__main__":
    main()
```
